```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#read the csv dataset STEP 1
movie_data = pd.read_csv('movie_dataset.csv')
movie_data.head()
```

```python
# to get column names or features STEP 2
movie_data.columns
```

```python
# select few features/ feature extraction STEP 3

selected_features = ['keywords','cast','genres','director']
```

```python
#Create a column in DF which combines all selected features STEP 4

for feature in selected_features:
    movie_data[feature] = movie_data[feature].fillna('')
```

```python
def combine_features(row):
    return row['keywords'] +" "+row['cast']+" "+row["genres"]+" "+row["director"]
movie_data["combined_features"] = movie_data.apply(combine_features,axis=1)
```

```python
movie_data["combined_features"].head()
```

```python
# STEP 5: count the no of similar words from combined_features column
# count matrix with count vetorizor class

from sklearn.feature_extraction.text import CountVectorizer

#create object of class CountVectorizer

count=CountVectorizer()

# fit_transform method is used to create count matrix

count_matrix = count.fit_transform(movie_data["combined_features"])
```

```python
count_matrix.toarray()
```

```python
#Compute the Cosine Similarity based on the count_matrix STEP 6
#cosine_similarity method is used for this purpose
# import library
from sklearn.metrics.pairwise import cosine_similarity
cos_sim = cosine_similarity(count_matrix)
```

```python
# input from user
user_choice = "Spider-Man 3"
```

```python
def get_title_from_index(index):
	return movie_data[movie_data.index == index]["title"].values[0]

def get_index_from_title(title):
	return movie_data[movie_data.title == title]["index"].values[0]
```

```python
# STEP 7 get index of avatar or from its title
get_movie_index = get_index_from_title(user_choice)
```

```python
similar_movies = list(enumerate(cos_sim[get_movie_index]))
```

```python
# STEp 8 sort the lists in decresing order

sorted_list_of_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)
```

```python

# print the complete list on decresing order
for movie in sorted_list_of_movies:
    print(get_title_from_index(movie[0]))
```

```python
# print 50 most similar movies

i = 0

for movie in sorted_list_of_movies:
    print(get_title_from_index(movie[0]))
    i =i+1
    if i>50:
        break
```
